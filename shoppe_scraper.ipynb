```python
type = 1
submit_time = datetime.now()  # Initialize to the current datetime
comment_list = []

# Run the loop until the condition is met
while type <= 5:
    offset = 0    
    while offset < 1000:
        try:
            # Make a request to the Shopee API
            shopee_request = requests.get(f'https://shopee.co.id/api/v4/seller_operation/get_shop_ratings_new?limit=100&offset={offset}&shopid=195455930&type={type}&userid=195458863')
            print(f"Date: {submit_time}\n Req: {shopee_request.status_code}\n Offset: {offset}\n Rating: {type}")

            # Check if the API returned a valid response
            if shopee_request.status_code != 200:
                print(f"API not 200. Status code: {shopee_request.status_code}")
                break

            req_as_json = shopee_request.json()

            # Safely extract comment data
            comment_data = req_as_json['data']['items']
            if not comment_data:  # Exit if no comments are found
                print("No more comments to process.")
                break

            # Process each comment
            for comment in comment_data:
                shop_id = comment.get('shopid', None)
                buyer_id = comment.get('userid', None)
                item_id = comment.get('itemid', None)
                submit_time = datetime.fromtimestamp(comment.get('submit_time', 0))
                rating_star = comment.get('rating_star', None)
                comment_buyer = comment.get('comment', None)

                comment_list.append(
                    {'Shop ID': shop_id,
                    'Buyer ID': buyer_id,
                    'Item ID': item_id,
                    'Time': submit_time,
                    'Rating': rating_star,
                    'Comment': comment_buyer})

            # Increment offset for the next API request
            offset += 100
            print(f"Offset updated to: {offset}")

            # Time delay to avoid being rate-limited
            time.sleep(5)

        except KeyError as e:
            print(f"KeyError encountered: {e}")
            break
        except Exception as e:
            print(f"Unexpected error: {e}")
            break
        
    type += 1

# Convert the collected data into a DataFrame and save as CSV
if comment_list:  # Only create a DataFrame if there's data
    comment_df = pd.DataFrame(comment_list)
    output_file = "../Shopee_scrapper/somethinc_comments.csv"
    comment_df.to_csv(output_file, index=False, encoding='utf-8')
    print(f"DataFrame created and saved as '{output_file}' successfully!")
else:
    print("No data collected. CSV file not created.")
```
